# a) Does the data satisfy Normality?

```r
shapiro.test(data$vital.capacity)
```

- P value > 0.05
- Failed to reject the null Hypothesis

- That means, Data is normally distrubuted.

# Conduct a hypothesis test to determine if the mean vital capacity for workers in the cadmium industry is significantly different than 4.7 liters.


```r
t.test(data$vital.capacity, mu=4.7)
```

- H0: Mean = 4.7
- Ha: mean not equla 4.7
- We cannot reject the null hypothsis that means cadmium industry is not significantly different than 4.7 liters

```r
boxplot(data$vital.capacity)
```

- We can clearly see here, Mean of vital capcity approx equal to 4.7 so we can conclude that mean = 4.7 not significantly 
- different than 4.7 

```r
t.test(data$vital.capacity~data$group)
```

- We will reject the NUll hypothesis in favor of alternative hypothesis
- difference between mean is not equal to zero
- we can clearly see the same result in the boxplot as well.
