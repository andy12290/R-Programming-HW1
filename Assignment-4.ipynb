# T test 

```r
# t test
y = data$score
x = data$gender
t.test(y~x)
```

- P value is < 0.05 that means we have to reject the nulll Hypothesis in favor of alternative hypothesis.
- that means, The mean difference between male and female not equal to zero
- YES. RESULT STAT SIGNIFICANT.
- Male getting more score than female
