```r
xvar <- c(1.59,3.51,2.51,1.47,2.22,2.97,1.75,3.05,4.71,2.54)
yvar <- c(9.3,14.8,9.7,6.5,11.9,12.1,10.3,13.6,18.8,11.1)
data <- data.frame(xvar,yvar)
ggplot(dat, aes(x=xvar, y=yvar)) +
    geom_point(shape=1) +    # Use hollow circles
    geom_smooth(method=lm)   # Add SLR line
                             # (default 95% CI)
```
